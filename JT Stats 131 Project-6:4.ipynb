## Top Grossing Actresses/Actors: Gender Differences

```python
#reads in new dataframe with info on top 10 grossing actors/actresses of all time
top_grossing_actors = pd.read_excel('top_grossing_actors.xlsx')
top_grossing_actors.head()
```

```python
# plot of top grossing actors. size of dot = total gross amount 
sizes = [10, 60, 90, 130, 200, 250] 
marker_size = pd.cut(top_grossing_actors['Total Gross (imdb)']/10000000, [0, 250, 300, 350, 400, 450, 500], labels=sizes) 
sns.lmplot('Total Movie Revenue','Number of Movies', data=top_grossing_actors, hue='Gender', fit_reg=False, scatter_kws={'s':marker_size})
plt.title('Num. Movies vs. Total Movie Revenue for Top Grossing Actors/Actresses')
plt.show()
```

```python
# plot of top grossing actors. size of dot = total movie revenue
sizes = [30,100,200,500] 
marker_size = pd.cut(top_grossing_actors['Total Movie Revenue']/100000000, [0, 50, 100, 150, 200], labels=sizes) 
sns.lmplot('Total Gross (imdb)','Number of Movies', data=top_grossing_actors, hue='Gender', fit_reg=False, scatter_kws={'s':marker_size})
plt.title('Num. Movies vs. Gross Revenue (imdb) for Top Grossing Actors/Actresses')
plt.show()
```

```python
# plot of top grossing actors. size of dot = number of movies
sizes = [30, 50, 100, 150, 200, 250, 300, 400, 500] 
marker_size = pd.cut(top_grossing_actors['Number of Movies'], [0, 20, 25, 30, 35, 40, 45, 50, 60, 90], labels=sizes) 
sns.lmplot('Total Gross (imdb)','Total Movie Revenue', data=top_grossing_actors, hue='Gender', fit_reg=False, scatter_kws={'s':marker_size})
plt.title('Total Movie Revenue vs. Total Gross (imdb) for Top Grossing Actors/Actresses')
plt.show()
```

## In-series vs. not in series Analysis

```python
movies['in_series'] = movies.belongs_to_collection.isna() == False
movies.head()#adds column of booleans for if movie is from a series or not
sns.violinplot(x='in_series', y='revenue', data=movies)
plt.title("Revenue for Movies In a Series vs. Not in a Series")
plt.show()
```

```python
sns.boxplot(x='in_series', y='revenue', data=movies)
plt.title("Revenue for Movies In a Series vs. Not in a Series")
plt.show()
```

```python
movies.groupby('in_series').revenue.mean() #highger revenue for movies in a series
```

```python
from sklearn import linear_model
```

```python
#Logistic Model to predict whether or not movie is in a series
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

logmodel = LogisticRegression()
movies2 = movies[['revenue','in_series']]
movies2 = movies2.dropna()

X_train, X_test, y_train, y_test = train_test_split(movies2.revenue, movies2.in_series, test_size=0.25)

#movies2['revenue_small'] = movies2.revenue/10000000

#X_train = movies2.revenue_small.iloc[:5000].values.reshape(-1,1)
#y_train = movies2.in_series.iloc[:5000]
#X_test = movies2.revenue_small.iloc[5001:].values.reshape(-1,1)
logmodel.fit(X_train.values.reshape(-1,1),y_train)
predictions = logmodel.predict(X_test.values.reshape(-1,1))
print(predictions)
sum(predictions == y_test)/len(predictions) #decent accuracy rate!
```
